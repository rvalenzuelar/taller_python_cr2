### Un ejemplo aplicado


* Pandas puede leer directamente desde archivos CSV (y muchos otros)

```python
cr2 = pd.read_csv('cr2_prDaily_2017.txt')
```

```python
cr2
```

* Le damos algunos argumentos para hacer mejor el parsing

```python
cr2 = pd.read_csv('cr2_prDaily_2017.txt',header=0, index_col=0, na_values=-9999)
```

```python
cr2
```

* Separamos la cabecera de los datos

```python
# selecciona hasta la fila 13 (indexing es exclusivo) 
# y todas las columnas
head = cr2.iloc[:14, :]

# selecciona desde la fila 14 y todas las columnas
# definimos el tipo de datos que contiene el DataFrame 
# (e.g. todos float)
data = cr2.iloc[14:, :].astype(float)  
```

```python
head
```

```python
data
```

```python
# seleccionamos una fila del header
head.loc['cantidad_observaciones']
```

```python
nobs = head.loc['cantidad_observaciones'].astype(int)
```

```python
nobs[nobs>10000]
```

* La primera columna de **`Series`** y **`DataFrames`** corresponden a indices


* En este caso, indices=ID_estacion y valor=num_obs

```python
# selecciona las ID de estaciones que tienen mas de 10.000
# observaciones y devuelve las series de tiempo
# con las observaciones (DataFrame)
obs_target = data[nobs[nobs>10000].index]
```

```python
obs_target
```

```python
obs_target.index
```

* En este momento los indices son **`strings`**. Los convertimos a **`datetime`** antes de graficar


* Importamos la funcion **`datetime`** desde el modulo **`datetime`**

```python
from datetime import datetime
```

* **`map`** aplica una funcion (`lambda`) por cada item en la lista (`obs_target.index`)


* **`lambda`** es una funcion anonima, puede contener otras funciones


* **`datetime.strptime`** convierte un `string` en un datetime

```python
index_datetime = map(lambda x: datetime.strptime(x,'%Y-%m-%d'), obs_target.index)
```

```python
index_datetime
```

```python
# cambia el antiguo indice (string) por datetime
obs_target.index = index_datetime
```

* Una de las funcionalidades de Pandas es crear estadistica agregada


* Una forma de hacer es usando la funcion **`resample`**

```python
fig,ax = plt.subplots()
estacion = nombre[nombre.str.contains('Carriel')]
rain = obs_target[estacion.index].astype(float)
rain.columns = estacion.values
rain = rain.dropna()  # elimina filas con NaN
freq = 'M'  # mensual
rain_M = rain.resample(freq).sum()  # suma mensual
rain_M.plot(ax=ax)
plt.show()
```

```python
fig,ax = plt.subplots()
estacion = nombre[nombre.str.contains('Carriel')]
rain = obs_target[estacion.index].astype(float)
rain.columns=['Carriel Sur']
rain = rain.dropna()  # elimina filas con NaN
freq = 'A' # anual
rain_Y = rain.resample(freq).sum()  # suma anual
rain_Y.plot(marker='s', markerfacecolor='none', ax=ax)
plt.show()
```

* Si queremos ver la anomalia

```python
rain_mean = rain_Y.mean()
```

```python
rain_anomaly = rain_Y - rain_mean
```

```python
fig,ax = plt.subplots()
rain_anomaly.plot(ax=ax)
plt.show()
```

* Podemos mejorar la visualizacion con la funcion **`fill_between`** de **`matplotlib`**

```python
fig,ax = plt.subplots()
rain_anomaly.plot(ax=ax, color='gray')
x = rain_anomaly.index
y1 = rain_anomaly.values.flatten()
y2 = [0] * len(rain_anomaly)
ax.fill_between(x, y1, y2, where=y2 >= y1, 
                facecolor='coral', interpolate=True)
ax.fill_between(x, y1, y2, where=y2 <= y1, 
                facecolor='steelblue', interpolate=True)
plt.show()
```

# Xarray

* Xarray esta inspirado en Pandas pero orientado a datos multidimensionales


* La estructura de los datos es similar a netCDF


* Asi como en Pandas hay **`Series`** y **`DataFrames`**, en **`Xarray`** hay **`DataArray`** y **`Dataset`**

```python
import xarray as xr
```

```python
da_pw = xr.open_dataarray('X200.9.97.140.288.12.52.29.nc')  # DataArray: 1 sola variable
```

```python
da_pw
```

* Seleccionamos un indice en la coordenada del tiempo usando la funcion **`isel`**


* Creamos la figura usando la funcion **`plot`**

```python
fig, ax = plt.subplots()
da_pw.isel(time=0).plot(ax=ax)
plt.show()
```

* La figura anterior muestra una escala con valores negativos


* Sabemos que el agua precipitable es una variable > 0


* Podemos filtrar valores negativos usando NaN

```python
import numpy as np
```

```python
da_pw.values[da_pw.values<0] = np.nan
```

```python
fig, ax = plt.subplots()
da_pw.isel(time=0).plot(ax=ax)
plt.show()
```

* En vez de un indice, con la funcion **`sel`** podemos usar un valor de la coordenada el tiempo para hacer la seleccion

```python
fig, ax = plt.subplots()
da_pw.sel(time=datetime(2017,10,3,0)).plot(ax=ax)
plt.show()
```

```python
da_mslp = xr.open_dataarray('X200.9.97.140.288.12.55.8.nc')
```

```python
fig, ax = plt.subplots()
da_mslp.isel(time=0).plot(ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots()
da_mslp.isel(time=0).plot(ax=ax,levels=11)
plt.show()
```

* Existe una variedad de tablas de colores
https://matplotlib.org/examples/color/colormaps_reference.html

```python
fig, ax = plt.subplots()
da_mslp.isel(time=0).plot(ax=ax,levels=11,cmap='inferno')
plt.show()
```

```python
fig, ax = plt.subplots()
da_mslp.isel(time=0).plot(ax=ax,levels=11,cmap='inferno_r')
plt.show()
```

```python
fig, ax = plt.subplots()
da_mslp.isel(time=0).plot.contourf(ax=ax,levels=11,cmap='inferno')
plt.show()
```

```python
fig, ax = plt.subplots()
da_mslp.isel(time=0).plot.contourf(ax=ax,levels=11,cmap='plasma')
plt.show()
```

* Un dataset contiene mas de una variable


* Las variables comparten dimensiones y coordenadas


* Podemos crear Datasets haciendo un **`merge`** de DataArrays

```python
ds = xr.merge([da_pw, da_mslp])
```

```python
ds
```

* Podemos calcular directamente estadistica agregada

```python
ds_daily = ds.resample('D','time',how='mean')
```

```python
fig,ax = plt.subplots()
ds['pr_wtr'].isel(time=0).plot(ax=ax)
plt.show()
```

```python
fig,ax = plt.subplots()
ds_daily['pr_wtr'].isel(time=0).plot(ax=ax)
plt.show()
```

* Podemos ver la evolución por día usando argumentos extras (x, y, col, col_wrap)

```python
ds_daily['pr_wtr'].plot(x='lon', y='lat', col='time', col_wrap=3, vmin=0, vmax=30)
plt.show()
```

```python
ds_daily['slp'].plot(x='lon', y='lat', col='time', col_wrap=3)
plt.show()
```

* Por defecto xarray crea una imagen (QuadMesh)


* Podemos seleccionar otro tipo de visualizacion (e.g. contourf)

```python
ds_daily['slp'].plot.contourf(x='lon', y='lat', col='time', col_wrap=3)
plt.show()
```
